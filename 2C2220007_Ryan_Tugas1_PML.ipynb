### Melakukan import library yang dibutuhkan untuk menemukan persamaan regresi

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

### Menginput data dalam berbentuk dictionary serta menyimpannya dalam variabel data

```python
data = {
    'xi': [3.2, 3.4, 3.8, 3.7, 4.1, 3.9, 3.3, 4, 3.8, 15],
    'yi': [1.5, 1.3, 0.9, 0.8, 0.7, 0.9, 1.4, 0.7, 0.9, 1.2]
}
```

### Mengubah bentuk data menjadi array numpy serta melakukan reshape

```python
xi = np.array(data['xi']).reshape(-1, 1)
yi = np.array(data['yi'])
```

### Membuat model Linear regression dengan fungsi "LinearRegression()" dan fungsi "fit()"

```python
# Membuat model regresi linear
model = LinearRegression()
model.fit(xi, yi)
```

### Mencari slope (b) dan intercept (a) dari model

```python
# Mencetak koefisien (slope) dan intercept dari model
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope}, Intercept: {intercept}")
```
